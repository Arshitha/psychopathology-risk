```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

path_to_csv= "/shared/project-psychopathology-risk/outputs/LMM/LMM_cohend.csv"
df= pd.read_csv(path_to_csv ,index_col=0)

cohens = sns.heatmap(df, cmap="YlGnBu")
plt.show

figure = cohens.get_figure()
figure.savefig('cohens_heatmap.png', dpi = 400)
```
